```python
import numpy                       
from matplotlib import pyplot                 
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16


Vmax=80 #km/hr

L=11    #km
DenMax=250 #cars/km
nx=51
dt=0.001 #hours
dx=L/(nx)
nt=int((3/60)/dt)

x = numpy.linspace(0,L,nx)
rho0 = numpy.ones(nx)*10
rho0[10:20] = 50
print(nt)
u = rho0
```

```python
#V=Vmax*(1-Den/DenMax)
#F[i]=Vmax*un[i]*(1-un[i]/DenMax)

F=numpy.zeros(nx)
v=F
v[0]=Vmax*(1-rho0[0]/DenMax)
F[0]=v[0]*rho0[0]

for n in range(1,nt):  
    un = u.copy()    
    for i in range(1,nx):
        v[i]=Vmax*(1-u[i]/DenMax)
        F[i]=v[i]*u[i]
        u[i] = un[i]-dt/dx*(F[i]-F[i-1])
       
        
pyplot.plot(x, u, color='#003366', ls='--', lw=3)
pyplot.ylim(0,60);
```

```python
import numpy                       
from matplotlib import pyplot                 
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16


Vmax=136 #km/hr

L=11    #km
DenMax=250 #cars/km
nx=51
dt=0.001 #hours
dx=L/(nx)
nt=int((3/60)/dt)

x = numpy.linspace(0,L,nx)
rho0 = numpy.ones(nx)*20
rho0[10:20] = 50
print(nt)
u = rho0
#-----------------------------------------------



F=numpy.zeros(nx)
v=F
v[0]=Vmax*(1-rho0[0]/DenMax)
F[0]=v[0]*rho0[0]

for n in range(1,nt):  
    un = u.copy()    
    for i in range(1,nx):
        v[i]=Vmax*(1-u[i]/DenMax)
        F[i]=v[i]*u[i]
        u[i] = un[i]-dt/dx*(F[i]-F[i-1])
       
        
pyplot.plot(x, u, color='#003366', ls='--', lw=3)
pyplot.ylim(0,60);
#---------------------------------------------------------------

v=numpy.zeros(nx)
for i in range(nx):
    v[i]=Vmax*(1-u[i]/DenMax)
    
print(v)
pyplot.plot(x, v, color='#003366', ls='--', lw=3)
pyplot.ylim(0,100);



sum=0


for i in range(nx):
    sum=sum+v[i]  

aver=sum/nx
print(aver)
aver=aver*1000/3600
print(aver)

```
